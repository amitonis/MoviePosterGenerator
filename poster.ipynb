```python
from diffusers import DiffusionPipeline
import torch

#Cleat GPU memory
torch.cuda.empty_cache()

model_id = "stabilityai/stable-diffusion-2-1-base"  # Low LEVEL Models("stabilityai/stable-diffusion-2-1-base" or "hakurei/waifu-diffusion")

#Using very low settings since insufficent VRAM
pipe = DiffusionPipeline.from_pretrained(
    model_id, 
    torch_dtype=torch.float32,  #FP16 for lower VRAM usage
    use_safetensors=True
)

# Activating GPU to run with the models for effiecent process
pipe.to("cuda")

# Memory optimizations
pipe.enable_attention_slicing()
pipe.enable_sequential_cpu_offload()

# PROMPT AND SIZE OF THE IMAGE
prompt = ""
image = pipe(prompt, width=720, height=720).images[0]  

image.save("lightweight_poster.png")
image.show()
```
